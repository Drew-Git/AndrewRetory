```python
from gensim.models import word2vec
from konlpy.tag import Twitter
from bs4 import BeautifulSoup
from urllib.request import urlopen
import re
```

```python
# 크롤링할 url
url = 'http://www.mediaus.co.kr/news/articleView.html?idxno=133389'
# 웹페이지 접속 후 결과 받기
res = urlopen(url) # 결과받고 # 클롤링에서 제일먼저 해야할 중요한건 내가 뽑아낼 데이터가 어느 태그에있는지
html = BeautifulSoup(res, 'html.parser')
article_body = html.find(id='articleBody').text
```

```python
article_body = re.sub('\.','\n', re.sub('\n','', article_body))
print(article_body)
```

```python
lines = article_body.split('\n') # 자른 후 형태소 분리 한다음에 다시 붙이기
twitter = Twitter()
result = []
for line in lines:
    news_analyze = twitter.pos(line, stem=True, norm=True)
    r = []
    for word in news_analyze:
        if word[1] not in ['Josa','Eomi','Punctuation']:
            r.append(word[0])
    rl = (" ".join(r)).strip()
    result.append(rl)
    print(rl)
```

```python
def txt_data_save(filename, data, encoding='utf-8'):
    with open(filename, 'w', encoding=encoding) as fp: # 배열에 들어있는거 한꺼번에 문자열로 처리하기
        fp.write('\n'.join(data))
```

```python
txt_data_save(filename='./data/ag_soccer.data', data=result) # 파일 저장하기
```

```python
data = word2vec.LineSentence('./data/ag_soccer.data') # LineSentence 문자열을 불러오겠다
# data를 집어넣어서 모델을 만듦. # 1번 등장하는 단어는 사용하지 않겠다 2개 이상 등장한 문자만 사용하곘다.
model = word2vec.Word2Vec(data, size=100, window=10, hs=1, min_count=2,sg=1) # size 매개변수가 중요함 
    #(차원수를 의미 하나의 단어에 대해서 몇차원에 벡터 행렬을 만들건지) 만다고 좋은것도 아니고 데이터가 많아질 수록 사이즈를 키워주면 됨 
    
model.save('./model/ag_soccer,model')
```

```python
model.most_similar(positive=['손흥민','조현우','슛']) # 연관된것들이 나옴 # 84%정도 연관이 있다
```

```python
model.most_similar(positive=['손흥민','골'],negative=['프리킥'])
```
